```python
# 연도별 정보통신방송기기의 전지역 규모 합계

idx = np.arange(3)
w = 0.15
ITrevenue1 = pd.DataFrame()
for year in ['2019', '2020', '2021']:
    df = ITrevenue.xs('소계', level = 2)[(year, '합계')].xs('정보통신방송기기', level = 0)
    df.drop('소계', inplace = True)
    ITrevenue1 = pd.concat([ITrevenue1, df], axis = 1, ignore_index = True)

ITrevenue1.columns = ['2019', '2020', '2021']
ITrevenue1 = ITrevenue1.T

plt.figure(figsize = (10, 5))
plt.title('<연도별 정보통신방송기기의 전지역 규모 합계>', size = 20)
plt.xlabel('연도', size = 20)
plt.ylabel('규모 (단위:조)', size = 20)
plt.bar(idx-2*w, ITrevenue1['전자부품'] / 10000, width = w, label = '전자부품')
for i, v in enumerate(zip([0, 1, 2], ITrevenue1['전자부품'].values)):
         plt.text(i - 2*w, v[1]/10000, round(v[1]/10000), fontsize = 12, color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.bar(idx-w, ITrevenue1['컴퓨터 및 주변기기'] / 10000, width = w, label = '컴퓨터 및 주변기기')
for i, v in enumerate(zip([0, 1, 2], ITrevenue1['컴퓨터 및 주변기기'].values)):
         plt.text(i - w, v[1]/10000, round(v[1]/10000), fontsize = 12, color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.bar(idx, ITrevenue1['통신 및 방송기기'] / 10000, width = w, label = '통신 및 방송기기')
for i, v in enumerate(zip([0, 1, 2], ITrevenue1['통신 및 방송기기'].values)):
         plt.text(i, v[1]/10000, round(v[1]/10000), fontsize = 12, color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.bar(idx+w, ITrevenue1['영상 및 음향기기'] / 10000, width = w, label = '영상 및 음향기기')
for i, v in enumerate(zip([0, 1, 2], ITrevenue1['영상 및 음향기기'].values)):
         plt.text(i + w, v[1]/10000, round(v[1]/10000), fontsize = 12, color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.bar(idx+2*w, ITrevenue1['정보통신응용기반기기'] / 10000, width = w, label = '정보통신응용기반기기')
for i, v in enumerate(zip([0, 1, 2], ITrevenue1['정보통신응용기반기기'].values)):
         plt.text(i + 2*w, v[1]/10000, round(v[1]/10000), fontsize = 12, color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.xticks(idx, ['2019', '2020', '2021'], size = 20)
plt.legend(fontsize = 9)
plt.show()
```

```python
# 기기별 규모 변화
def change_in_sales(classi1, classi2):
    region_list = ['서울', '경기', '인천', '부산, 울산, 경남', '대구, 경북', '광주, 전남, 전북, 제주', '대전, 충남, 충북, 세종, 강원']
    plt.figure(figsize = (10, 5))
    i = 0
    for region in region_list:
        revenues = []
        for year in ['2019', '2020', '2021']:
            ITrevenue2 = ITrevenue.xs('소계', level = 2)[(year, region)]
            revenues.append(ITrevenue2.loc[classi1, classi2]/10000)    
        plt.plot(revenues, label = region_list[i])
        i += 1
    plt.style.use('ggplot')
    plt.xticks([0, 1, 2], ['2019', '2020', '2021'], size = 20)
    #plt.yticks([0, 1, 8, 10, 12], size = 20)
    plt.xlabel('연도', size = 20)
    plt.ylabel('매출액 (단위:조)', size = 20)
    plt.title(f'<{classi1} : {classi2}의 지역에 따른 연별 규모 변화>')
    plt.legend()
    plt.show()
```

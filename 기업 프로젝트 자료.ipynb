```python
SALARYTIME=pd.read_excel(file5, index_col=[0,1,2], header=[0,1])
NEWSALARY=pd.read_excel(file6, index_col=[0,1,2], header=[0,1])
```

```python
NEWSALARY
```

```python
SALARYTIME = SALARYTIME.xs('J.정보통신업(58~63)', level=1).xs('중소규모(5~299인)', level=1)
SALARYTIME
```

```python
NEWSALARY = NEWSALARY.xs('J. 출판，영상，방송통신및정보서비스업(58~63)', level=1).xs('중소규모(5~299인)', level=1)
NEWSALARY
```

```python
SALARYS = pd.concat([NEWSALARY, SALARYTIME], axis=1, join='inner')
SALARYS
```

```python
SALARYS[('2019', '상용근로일수 (일)')]
```

```python
SALARYS[('2019', '상용월급여액 (원)')]
```

```python
SALARYS = SALARYS.drop(['2017', '2018'], axis=1, level=0)
SALARYS
```

```python
SALARYS = SALARYS.astype(float)
```

```python
for i in range(2019, 2022):
    SALARYS[(f'{i}', '단위 일수당 상용 월급여')]=\
    round(SALARYS[(f'{i}', '상용월급여액 (원)')]/SALARYS[(f'{i}', '상용근로일수 (일)')], 2)
SALARYS = SALARYS.sort_index(axis=1, level=[0], ascending=[True])
SALARYS
```

```python
SALARYS_2019 = SALARYS.xs('2019', axis=1, level=0)
plt.figure(figsize=(16, 6))
plt.bar(SALARYS_2019.index, SALARYS_2019['상용근로일수 (일)'])
```

```python
SALARYS_2019 = SALARYS.xs('2019', axis=1, level=0)
SALARYS_2020 = SALARYS.xs('2020', axis=1, level=0)
SALARYS_2021 = SALARYS.xs('2021', axis=1, level=0)
SALARYS_list = [SALARYS_2019, SALARYS_2020, SALARYS_2021]
```

```python
width = 0.25
multiplier = 0
x = np.arange(len(SALARYS_list[0].index))
# color_list = ['#4374D9', '#FF5A5A', '#2F9D27', '#3DB7CC', '#FFBB00', '#9FC93C']

fig, ax = plt.subplots(figsize=(20,8))

for idx, data in enumerate(SALARYS_list):
    offset = width * multiplier
    rects = ax.bar(x + offset, round(SALARYS_list[idx]['상용월급여액 (원)']/10000), width=width,
                   label=f'{2019+idx}년 상용월급여액')
    ax.bar_label(rects, padding=3)
    multiplier += 1
ax.set_xlabel('지역', fontsize=15)
ax.set_ylabel('상용월급여액(만원)', fontsize=15)
ax.set_xticks(x + width, SALARYS_list[0].index)
ax.set_yticks(range(0, 501, 50))
ax.legend(loc='upper right', ncol=3)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.suptitle('2019~2021년 지역별 상용월급여액', fontsize=20)
plt.show()
```

```python
width = 0.25
multiplier = 0
x = np.arange(len(SALARYS_list[0].index))
# color_list = ['#4374D9', '#FF5A5A', '#2F9D27', '#3DB7CC', '#FFBB00', '#9FC93C']

fig, ax = plt.subplots(figsize=(20,8))

for idx, data in enumerate(SALARYS_list):
    offset = width * multiplier
    rects = ax.bar(x + offset, round(SALARYS_list[idx]['단위 일수당 상용 월급여']/10000,1), width=width,
                   label=f'{2019+idx}년 단위 일수당 상용 월급여')
    ax.bar_label(rects, padding=3)
    multiplier += 1
ax.set_xlabel('지역', fontsize=15)
ax.set_ylabel('단위 일수당 상용 월급여(만원)', fontsize=15)
ax.set_xticks(x + width, SALARYS_list[0].index)
ax.set_yticks(range(0, 25, 2))
ax.legend(loc='upper right', ncol=3)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.suptitle('2019~2021년 지역별 단위 일수당 상용 월급여', fontsize=20)
plt.show()
```
